# Practice 9. Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from requests import get
from io import StringIO
url = 'https://raw.githubusercontent.com/dj-kostya/NSU_PYTHON_2_Semestr_Practices/master/TMP_DATA/company_sales_data.csv'
r = get(url)
data_csv = StringIO(r.text)
DataFrame = pd.read_csv(data_csv)
DataFrame
```

## Task 1

```python
plt.figure(figsize=(8, 6))
plt.plot(DataFrame['month_number'], DataFrame['total_profit'])
plt.ylabel('Total profit')
plt.xlabel('Month Number')
plt.title('Company profit per month')
plt.show()
```

## Task 2

```python
plt.figure(figsize=(8, 6))
plt.plot(DataFrame['month_number'], DataFrame['total_profit'], 
         linestyle='--', marker='o', color='r',  markerfacecolor='k', linewidth=3, label='Profit data of last year')
plt.ylabel('Sold units number')
plt.xlabel('Month Number')
plt.title('Company sales data of last year')
plt.legend()
plt.show()
```

## Task 3

```python
labels = {
    'Face cream': 'facecream',
    'Face wash': 'facewash',
    'Tooth Paste': 'toothpaste',
    'Bathing Soap': 'bathingsoap',
    'Shampoo': 'shampoo',
    'Moisturizer': 'moisturizer'
}
plt.figure(figsize=(8, 6))
for label in labels:
    plt.plot(DataFrame['month_number'], DataFrame[labels[label]], marker='o', label=f'{label} Sales Data')
plt.ylabel('Sales units in number')
plt.xlabel('Month Number')
plt.title('Sales data')
plt.legend()
plt.show()
```

## Task 4

```python
plt.figure(figsize=(8, 6))
plt.scatter(DataFrame['month_number'], DataFrame['toothpaste'], label='Tooth Paste Sales Data')
plt.grid(linestyle='--')
plt.ylabel('Number of units sold')
plt.xlabel('Month Number')
plt.title('Tooth Paste Sales Data')
plt.legend()
plt.show()
```

## Task 5

```python
ax = DataFrame.plot(x='month_number', y=['facecream', 'facewash'], kind='bar', rot=0)
ax.legend(['Face cream Sales Data', 'Face wash Sales Data'])
ax.grid(linestyle='--')
plt.ylabel('Sales units in number')
plt.xlabel('Month Number')
plt.title('Face wash and Face cream Sales Data')
plt.show()
```

## Task 6

```python
ax = DataFrame.plot(x='month_number', y='bathingsoap', kind='bar', rot=0, legend=False)
ax.grid(linestyle='--')
plt.ylabel('Sales units in number')
plt.xlabel('Month Number')
plt.title('bathingsoap Sales Data')
plt.show()
```

## Task 7

```python
plt.figure(figsize=(8, 6))
plt.hist(DataFrame['total_profit'],bins=9, label='Profit data', range=(150000,400000))
plt.ylabel('Actual Profit in dollar')
plt.xlabel('Profit range in dollar')
plt.title('Profit data')
plt.show()
```

## Task 8

```python
sum_ = [sum(DataFrame[labels[label]]) for label in labels]
plt.figure(figsize=(8, 6))
plt.pie(sum_,  labels=labels, autopct='%1.1f%%')
plt.legend()
plt.title('Sales data')
plt.show()
```

## Task 9

```python
plt.subplot(211)
plt.plot(DataFrame['month_number'], DataFrame['bathingsoap'], marker='o',color='k')
plt.title('Sales data of Bathingsoap')
plt.xticks([])
plt.subplot(212)
plt.plot(DataFrame['month_number'], DataFrame['facewash'], marker='o', color='r')
plt.title('Sales data of a facewash')
plt.ylabel('Sold units in number')
plt.xlabel('Month Number')
plt.show()
```

## Task 10

```python

ax=DataFrame.plot.area(y=labels.values(),stacked=True, figsize=(8,6))
ax.legend([f'{label} Sales Data' for label in labels])
plt.ylabel('Sales units in number')
plt.xlabel('Month Number')
plt.title('All product sales data using stack plot')
plt.show()
```
